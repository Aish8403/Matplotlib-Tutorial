```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

Loading Scikit-Learn Datasets as Pandas DataFrames
Scikit-Learn provides multiple datasets. Let’s load one of them into a Pandas DataFrame.

Example: Load the Iris dataset

```python
# Load the dataset
iris = datasets.load_iris()

# Convert to Pandas DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target variable
df['target'] = iris.target

# Display the first 5 rows
df.head()
```

Exploring the Dataset with Pandas

```python
#Basic Information
df.info()   # Summary of dataset
df.describe()  # Statistical summary
df['target'].value_counts()  # Count of each class in the target column
```

```python
df.isnull().sum()  # Check for missing values
```

Data Visualization with Matplotlib
Now, let’s visualize some relationships using Matplotlib.

```python
#Histogram

plt.figure(figsize=(8, 5))
df['sepal length (cm)'].hist(bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')
plt.show()
```

```python
#Scatter Plot

plt.figure(figsize=(8, 5))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['target'], cmap='viridis', alpha=0.7)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
plt.colorbar(label="Target Class")
plt.show()
```

```python
#Box Plot

plt.figure(figsize=(8, 5))
df.boxplot(column=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
plt.title('Feature Distributions')
plt.show()
```

Use California Housing Dataset 

```python
from sklearn.datasets import fetch_california_housing

# Load dataset
california = fetch_california_housing()

# Convert to Pandas DataFrame
california_df = pd.DataFrame(california.data, columns=california.feature_names)

# Add target column (Median House Value)
california_df['MEDV'] = california.target

# Display first few rows
california_df.head()
```

```python
#Histogram of Median House Value

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(california_df['MEDV'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.title('Distribution of Median House Values in California')
plt.show()
```

```python
#Scatter Plot (House Value vs Median Income)

plt.figure(figsize=(8, 5))
plt.scatter(california_df['MedInc'], california_df['MEDV'], alpha=0.5, color='blue')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.title('House Value vs Median Income in California')
plt.show()
```

```python
#Correlation Heatmap

import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(california_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.ion()  # Enable interactive mode

for i in range(1, 6):
    plt.figure(figsize=(6, 4))
    plt.plot(np.random.rand(10), np.random.rand(10), marker='o', linestyle='--', label=f'Plot {i}')
    plt.title(f'Interactive Plot {i}')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.show()
    plt.pause(1)  # Pause for 1 second before the next plot

plt.ioff()  # Disable interactive mode
```

Explanation<br>
plt.ion() → Turns on interactive mode.<br>
Loop through 5 iterations → Creates multiple plots dynamically.<br>
plt.pause(1) → Adds a 1-second pause to visualize each plot before updating.<br>
plt.ioff() → Turns off interactive mode after execution.<br>

```python
import time

plt.ion()
fig, ax = plt.subplots()

x_data, y_data = [], []
for i in range(20):
    x_data.append(i)
    y_data.append(np.sin(i / 2))  # Simulating some changing data
    
    ax.clear()  # Clear the previous frame
    ax.plot(x_data, y_data, marker='o', linestyle='-', color='b')
    ax.set_title('Live Updating Line Plot')
    ax.set_xlabel('Time')
    ax.set_ylabel('Value')
    
    plt.draw()
    plt.pause(0.5)  # Pause for half a second to simulate live updates

plt.ioff()
plt.show()
```
